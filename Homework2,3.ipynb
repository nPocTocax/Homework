```python
import pandas as pd

df = pd.read_csv('/usr/local/lib/python3.10/dist-packages/pandas/io/Customers.csv', delimiter=';')

# Найти людей, у которых возраст больше 30 и доход меньше 30000
filtered_1 = df[(df['Age'] > 30) & (df['Income'] < 30000)]

# Найти людей, которые по профессии юристы и с опытом работы больше 5 лет
filtered_2 = df[(df['Profession'] == 'Lawyer') & (df['Work Experience'] > 5)]

print("Люди c возрастом больше 30 и доходом меньше 30000:")
print(filtered_1)

print("\nЛюди, которые юристы и имеют опыт работы больше 5 лет:")
print(filtered_2)
```
